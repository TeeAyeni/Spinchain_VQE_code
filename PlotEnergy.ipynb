# Energy against opt. steps

```python
# needed modules
from scipy.io import loadmat
import matplotlib.pyplot as plt  
import numpy as np
import statistics
```

## 2 particles on 4 sites

```python
# 2 particles on 4 sites
fileA = "XXZ_2_4_A_no_ent_BWC_nonoise"
fileB = "XXZ_2_4_B_no_ent_BWC_nonoise"
fileG = "XXZ_2_4_G_no_ent_BWC_nonoise"

# load data from files
dataA = loadmat(fileA)
dataB = loadmat(fileB)
dataG = loadmat(fileG)

# average energy (averaged over 20 trials)
avg_energy_dataA = dataA['avg_energy_trials'][0]
avg_energy_dataB = dataB['avg_energy_trials'][0]
avg_energy_dataG = dataG['avg_energy_trials'][0]

# an array of the exact ground state energy
data_len = np.size(avg_energy_dataA)
exact_energy = dataA['Exact GS energy'][0]
exact_energy_data = np.ones(data_len)*exact_energy

# average relative error 
rel_err_dataA = dataA['avg_relerr_trials'][0]
rel_err_dataB = dataB['avg_relerr_trials'][0]
rel_err_dataG = dataG['avg_relerr_trials'][0]
```

```python
dataA.keys()
```

```python
# plot of energy
plt.plot(avg_energy_dataA, '-.', label = "A", color='#E52B50')
plt.plot(avg_energy_dataB, '--', label = "B", color='#6EB5E7')
plt.plot(avg_energy_dataG, '-', label = "G", color='#68B87B', mfc="none")

plt.xscale('log')

# labels
plt.legend(loc ='lower left')
plt.xlabel("opt. steps")
plt.ylabel("$E$")

# save plot
figname = "XXZ_2_4_Energy"
plt.savefig(figname + ".eps", bbox_inches='tight')
```

```python
# plot of energy relative error
plt.plot(rel_err_dataA, '-.', label = "A", color='#E52B50')
plt.plot(rel_err_dataB, '--', label = "B", color='#6EB5E7')
plt.plot(rel_err_dataG, '-', label = "G", color='#68B87B', mfc="none")

# activate log scaling
plt.xscale('log')
plt.yscale('log')


# labels
plt.legend(loc ='lower left')
plt.xlabel("opt. steps")
plt.ylabel("Relative error")

# save plot
figname = "XXZ_2_4_err"
plt.savefig(figname + ".eps", bbox_inches='tight')
```
